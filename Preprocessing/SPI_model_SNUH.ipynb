```python
import numpy as np
import pandas as pd
from pyvital2 import arr
from pyvital2 import pleth_spi
import numpy as np
import scipy.stats as st
from scipy.optimize import curve_fit
import os, pickle

srate = 100

# x_test loading
x_test = np.load('../DL_model/dataset/x_test_comp.npz', allow_pickle=True)['arr_0']
# test set PPG 
x_test_pacu = x_test[0:15140,:,0]
x_test_preop = x_test[15140:,:,0]

# test set  caseids
filepath_test = pickle.load(open('../DL_model/filepath_test','rb'))
vital_list = np.array(pd.Series(filepath_test).str.split(',', n=1).tolist())[:,1]

file_spi_test_pacu = pickle.load(open('../DL_model/file_spi_test_pacu','rb'))
temps = np.array(pd.Series(file_spi_test_pacu).str.split(',', n=1).tolist())[:,1]
vital_test_pacu = [temp[:-3]+'vital' for temp in temps]


spi_y_pred = []

# pacu test data  spi 
for idx, pleth in enumerate(x_test_pacu):
    print('{}/{}...'.format(idx+1, len(x_test_pacu)), end='')
    
    
    # individual histogram    
    f_vital = vital_test_pacu[idx]
    hist_path = 'cache/histogram/'+f_vital
    if os.path.exists(hist_path):
        hist_hbi, hist_ppga = pickle.load(open(hist_path,'rb'))
    
    #   SPI  2    
    else:
        spi_y_pred.append(-1)
        continue
    
    spi_per_input = []
    
    #  HBI, PPGA      
    print('detecting peaks...',end='')
    minlist, maxlist = arr.detect_peaks(pleth, 100)

    for j in range(len(maxlist) - 1):
        # HBI, PPGA   
        hbi = (maxlist[j+1] - maxlist[j]) / srate * 1000
        ppga = pleth[maxlist[j+1]] - pleth[minlist[j]]

        #   histogram 
        if 240 < hbi < 2000:
            hist_hbi.learn(hbi)     
        else:
            if 0 < ppga < 100:
                hist_ppga.learn(ppga)
                continue
            continue

        if 0 < ppga < 100:
            hist_ppga.learn(ppga)
        else:
            continue

        # HBI PPGA norm
        hbi_perc = hist_hbi.percentile(hbi)*0.73 + group_hist_hbi_2.percentile(hbi)*0.27
        ppga_perc = hist_ppga.percentile(ppga)*0.73 + group_hist_ppga_2.percentile(ppga)*0.27

        # SPI 
        spi = 100 - (0.7*ppga_perc + 0.3*hbi_perc)
        spi_per_input.append(spi)

    # input(10)   hbi, ppga  
    if spi_per_input == []:
        spi_y_pred.append(-2)
        
    spi_y_pred.append(np.mean(spi_per_input)/10)
    print('done')
```

```python
# preop data  spi 
file_spi_test_preop = pickle.load(open('../DL_model/file_spi_test_preop','rb'))
temps = np.array(pd.Series(file_spi_test_preop).str.split(',', n=1).tolist())[:,1]
vital_test_preop = [temp[:-3]+'vital' for temp in temps]

spi_y_pred_preop = []
# preop test data  spi 
for idx, pleth in enumerate(x_test_preop):
    print('{}/{}...'.format(idx+1, len(x_test_preop)), end='')
    
    f_vital = vital_test_preop[idx]
    
    
    # individual histogram
    hist_path = 'cache/histogram/'+f_vital
    if os.path.exists(hist_path):
        hist_hbi, hist_ppga = pickle.load(open(hist_path,'rb'))
    
    #   SPI  2    
    else:
        spi_y_pred_preop.append(-1)
        continue
    
    spi_per_input = []
        
    #  HBI, PPGA      
    print('detecting peaks...',end='')
    minlist, maxlist = arr.detect_peaks(pleth, 100)

    hbi_list, ppga_list = [], []
    for j in range(len(maxlist) - 1):
        # HBI, PPGA   
        hbi = (maxlist[j+1] - maxlist[j]) / srate * 1000
        ppga = pleth[maxlist[j+1]] - pleth[minlist[j]]
        if 240 < hbi < 2000:
            hist_hbi.learn(hbi)
            hbi_list.append(hbi)
        else:
            if 0 < ppga < 100:
                hist_ppga.learn(ppga)
                continue
            continue

        if 0 < ppga < 100:
            hist_ppga.learn(ppga)
            ppga_list.append(ppga)    
        else:
            continue

        # HBI PPGA norm
        hbi_perc = hist_hbi.percentile(hbi)*0.73 + group_hist_hbi_2.percentile(hbi)*0.27
        ppga_perc = hist_ppga.percentile(ppga)*0.73 + group_hist_ppga_2.percentile(ppga)*0.27

        # SPI 
        spi = 100 - (0.7*ppga_perc + 0.3*hbi_perc)
        spi_per_input.append(spi)

    # input   hbi, ppga  
    if spi_per_input == []:
        spi_y_pred.append(-2)

    spi_y_pred_preop.append(np.mean(spi_per_input)/10)
    print('done')
```
