```python
import pickle
import pandas as pd

with open('filepath_test','rb') as f:
    prepro = pickle.load(f)

with open('../DL_model/filepath_test','rb') as g:
    DL = pickle.load(g)

print(len(prepro))
print(len(DL))

for i in prepro:
    if i not in DL:
        print(0)
        
test_list = prepro
print(test_list)
```

```python
caseid=[]
row=[]
for j in test_list:
    caseid.append(j.split(',')[2])
    row.append(j.split(',')[1])
caseid_v=[]
for k in caseid:
    caseid_v.append(k.replace('.csv','.vital'))

caseid = caseid_v
print(len(caseid))
```

```python

df_match = pd.read_csv('match_drug_arr_exclusion_20201110.csv',sep=',')

set_value = []
cnt =0
for x in range(len(df_match)):
    if df_match.loc[x,'path'] in caseid:
        set_value.append('1')
        cnt +=1
    else:
        set_value.append('0')

rowv = []        
count=0
for y in range(len(df_match)):
    if y in row:
        rowv.append('1')
        count +=1
    else:
        rowv.append('0')
df_match['set']=set_value
print(cnt)
print(count)
#for y in range(len(df_match)):
    #노트북 코드 가져올것
#df_match.to_csv('match_20201125_set.csv',sep=',')

print(df_match)
```
